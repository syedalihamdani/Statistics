```python
import numpy as np
import pylab
import scipy.stats as stats
```

```python
std_normal=np.random.normal(loc=0,scale=1,size=1000)
for i in range(1,101):
    print(i,np.percentile(std_normal,i))
```

```python
stats.probplot(std_normal,dist="norm",plot=pylab)
pylab.show()
```
