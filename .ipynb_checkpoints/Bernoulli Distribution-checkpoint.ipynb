```python
from scipy.stats import bernoulli
import seaborn as  sns
```

```python
data_bern=bernoulli.rvs(size=1000,p=0.6)
ax=sns.distplot(data_bern,kde=
               True,color='crimson',
               hist_kws={"linewidth":25,'alpha':1})
ax.set(xlabel='Bernoulli',ylabel='Frequency')
```

```python
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
```

```python
# Instance of Bernoulli distribution with parameter
```

```python
p=0.7
```

```python
bd=bernoulli(0.7)
```

```python
# Outcome of expreiment can take value as 0,1
```

```python
X=[0,1]
```

```python
# Create a bar plot; Note the usage of "pmf" function
```

```python
# to determine the probability of different values of random variable
```

```python
plt.figure(figsize=(7,7))
plt.xlim(-1,2)
plt.bar(X,bd.pmf(X),color='orange')
plt.title('Bernoulli Distribution (p=0.7)',fontsize='15')
plt.xlabel('Values of Random variable X(0,1)',fontsize='15')
plt.ylabel('Probability',fontsize='15')
plt.show()
```
